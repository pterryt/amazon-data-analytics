## Clean Data / Column Pruning

##### Drop null rows (2).

```python
df.dropna(inplace=True)
```

##### Prune unnecessary columns. Add auto-incrementing key.

```python
df.drop(columns=['discounted_price', 'about_product', 'img_link'], inplace=True)
df['unique_id'] = range(1, len(df) + 1)
```

##### Initial cleaning of numbers and product name.

```python
# Trim product_name to simply hold the brand name
df['product_name'] = df['product_name'].str.split(' ').str[0]
df.rename(columns={'product_name': 'brand'}, inplace=True)

# Remove signs and commas from numbers
df.discount_percentage = df['discount_percentage'].str.replace('%', '')
df.actual_price = df['actual_price'].str.replace('₹', '')
df.rating_count = df['rating_count'].str.replace(',', '')
df.actual_price = df['actual_price'].str.replace(',', '')
```

##### Remove inconsistent values / convert numeric types

**Let's find out how many ratings can't be converted.**

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
invalid_count = df['rating'].isna().sum()
print(f'Non-numeric ratings: {invalid_count}')
```

**Just one. We'll drop it.**

```python
df.dropna(subset=['rating'], inplace=True)
```

**Convert to numeric types now that the values are consistent.**

```python
df['actual_price'] = df['actual_price'].round().astype(float).astype(int)
df['discount_percentage'] = df['discount_percentage'].astype(int)
df['rating_count'] = df['rating_count'].astype(int)
df['rating'] = df['rating'].astype(float)
```

##### Validate Types

```python
df.info()
```

### Reduce category to the child most category and create a separate table for categories.

```python
categories = []
for row in df['category']:
    levels = row.split('|')
    for i, name in enumerate(levels):
        parent = levels[i - 1] if i > 0 else None
        categories.append({'name': name, 'parent': parent})

df_categories = pd.DataFrame(categories)
df_categories = df_categories.drop_duplicates(subset=['name', 'parent'])

# Reduce categories on main df
df['category'] = df['category'].str.split('|').str[-1]

# Preview main dataframe
print('\n')
df.head(5)
```

## Exploratory Data Analysis


### Question: Are products with higher discounts more likely to have higher sales (approximated by rating_count)?
Null Hypothesis (H₀): There is no relationship between the discount percentage and the rating count (no correlation).<br>
Alternative Hypothesis (H₁): There is a significant positive correlation between the discount percentage and the rating count (higher discounts lead to higher rating counts).

__A scatterplot is a good initial visualization here for seeing distribution and any correlations/patterns there may be.__

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(df['discount_percentage'], df['rating_count'], alpha=0.6, color='blue')

plt.xlabel('Discount Percentage (%)', fontsize=12)
plt.ylabel('Rating Count', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

**Let's categorize into bins and calculate the percentage increase from one bin to the next for better clarity.**

#### Break data into bins for discount percentage


```python
bins = [0, 20, 40, 60, 80, 100]
labels = ['0–20%', '21–40%', '41–60%', '61–80%', '81–100%']
df['discount_range'] = pd.cut(df['discount_percentage'], bins=bins, labels=labels)

summary = df.groupby('discount_range', observed=True)['rating_count'].agg(['mean', 'count'])
summary['percent_increase'] = summary['mean'].pct_change() * 100
summary = summary.reset_index()

from IPython.display import display
print('\n')
display(summary)
```

**At first glance it is difficult to see a relatiosnip.**

#### Run a Pearson correlation test to access the p-value

```python
import scipy.stats as stats

correlation, p_value = stats.pearsonr(df['discount_percentage'], df['rating_count'])
print(f"Pearson correlation: {correlation}")
print(f"P-value: {p_value}")
```

**High p-value.**
**But is the data suited for Pearson?**<br>
**Let's look at statistical descriptions of our variables.**

```python
desc1 = df['actual_price'].describe()
desc2 = df['rating_count'].describe()
desc3 = df['discount_percentage'].describe()
desc4 = df['rating'].describe()
result = pd.concat([desc1, desc2, desc3, desc4], axis=1)
result.columns = ['actual_price', 'rating_count', 'discount_percentage', 'rating']
result
```

**Large difference in IQR (interquartile range) in actual price and rating count reveal high skew.**

**We'll do OLS regression modeling, with model fit assessment, and assumption checks.**<br>
**Reasoning: While residual diagnostics show some assumptions of OLS are not fully met (normality of residuals and homoscedasticity)**<br>
**OLS regression can still provide insights into variable relationships. It also flags areas where alternative methods or transformations might be needed.**

```python
import statsmodels.api as sm

X = df[['discount_percentage', 'actual_price', 'rating']]
y = df['rating_count']
X = sm.add_constant(X)  # Add intercept
model = sm.OLS(y, X).fit()
print(model.summary())
```

#### Observations
**R-squared is very low suggesting that independent variables have very little explanatory power for predicting rating count.**
**Low adjusted R-squared reinforces this idea.**

**We have at least one independent variable that is stastically significant (F-stat 6.59) (p-value .000201).**

rating: Coefficient 16,490.0, standard error 3912.37, t-statistic 4.214, and p-value 0.000<br>
 * Highly significant, indicating that products with higher ratings tend to have higher rating counts. <br><br>

**Lastly, for the sake of experimentation we'll see if categorization of actual price yields any relationship.**

#### Break data into (5) quantile price bins

```python
# Create price bins based on quantiles
price_bins = [0] + list(df['actual_price'].quantile([0.2, 0.4, 0.6, 0.8]).values) + [df['actual_price'].max()]
price_labels = ['Quantile 1', 'Quantile 2', 'Quantile 3', 'Quantile 4', 'Quantile 5']

# Add a new column for price groups
df['price_group'] = pd.cut(df['actual_price'], bins=price_bins, labels=price_labels)

# Visualize data for all 5 groups
for group in price_labels:
    group_data = df[df['price_group'] == group]

    # Check if there is any data in the group
    if len(group_data) > 0:
        plt.figure(figsize=(8, 5))
        plt.scatter(group_data['discount_percentage'], group_data['rating_count'], alpha=0.6)
        plt.title(f'Actual Price {group} - ({len(group_data)})', fontsize=14)
        plt.xlabel('Discount Percentage (%)', fontsize=12)
        plt.ylabel('Rating Count', fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.show()
    else:
        print(f"No data available for {group}")
```

#### Conclusion
Graphical analysis fails to reveal patterns that suggest relation.<br>
The OLS results suggest that **discount percentage** has a weak relationship with **rating count**. <br>
Even so, distribution of variables and potential outliers indicate that the data might need further cleaning and transformation before any solid conclusions can be drawn (e.g. log transformation).<br>
