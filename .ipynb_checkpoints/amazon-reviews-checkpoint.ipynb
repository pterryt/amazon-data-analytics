## Import Raw Data from CSV and Inspect Structure

##### Import data from CSV.

```python
import pandas as pd
csv_file = 'data/amazon.csv'
df = pd.read_csv(csv_file)
```

```python
print(df.isnull().sum()[df.isnull().sum() > 0])
```

##### Get row/column count and view types.

```python
df.info()
```

##### Inspect the contents.

```python
df.head()
```

##### Duplicate check.

```python
duplicate_counts = df.apply(lambda col: col.duplicated().sum())
print(duplicate_counts)
```

## Clean Data / Column Pruning

##### Drop null rows (2).

```python
df.dropna(inplace=True)
```

##### Prune unnecessary columns. Add auto-incrementing key.

```python
wdf = df.drop(columns=['discounted_price', 'about_product', 'img_link'])
wdf['unique_id'] = range(1, len(wdf) + 1)
```

##### Initial cleaning of numbers and product name.

```python
# Trim product_name to simply hold the brand name
wdf['product_name'] = wdf['product_name'].str.split(' ').str[0]
wdf.rename(columns={'product_name': 'brand'}, inplace=True)

# Remove signs and commas from numbers
wdf.discount_percentage = wdf['discount_percentage'].str.replace('%', '')
wdf.actual_price = wdf['actual_price'].str.replace('₹', '')
wdf.rating_count = wdf['rating_count'].str.replace(',', '')
wdf.actual_price = wdf['actual_price'].str.replace(',', '')
```

##### Remove inconsistent values / convert numeric types

**Let's find out how many ratings can't be converted.**

```python
wdf['rating'] = pd.to_numeric(wdf['rating'], errors='coerce')
invalid_count = wdf['rating'].isna().sum()
print(f'Non-numeric ratings: {invalid_count}')
```

**Just one. We'll drop it.**

```python
wdf.dropna(subset=['rating'], inplace=True)
```

**Convert to numeric types now that the values are consistent.**

```python
wdf['actual_price'] = wdf['actual_price'].round().astype(float).astype(int)
wdf['discount_percentage'] = wdf['discount_percentage'].astype(int)
wdf['rating_count'] = wdf['rating_count'].astype(int)
wdf['rating'] = wdf['rating'].astype(float)
```

##### Validate Types

```python
wdf.info()
```

### Reduce category to the child most category and create a separate table for categories.

```python
categories = []
for row in df['category']:
    levels = row.split('|')
    for i, name in enumerate(levels):
        parent = levels[i - 1] if i > 0 else None
        categories.append({'name': name, 'parent': parent})

df_categories = pd.DataFrame(categories)
df_categories = df_categories.drop_duplicates(subset=['name', 'parent'])

# Reduce categories on main df
wdf['category'] = wdf['category'].str.split('|').str[-1]

# Preview main dataframe
wdf.head(5)
```

```python
# Preview new table
print(df_categories.head(6))
```

### Create separate table for reviews

##### Understanding and cleaning user_id, review_id, review_title, review_content

**Count commas in comma separated fields to determine how many can be extracted.**

```python
def char_counts (column_name:str, char:str):
    count = wdf[column_name].apply(lambda x: x.count(char))
    percentage = (count.value_counts(normalize=True) * 100).sort_index()
    print(f"\n{char} count in {percentage.head(10)}")

char_counts('review_id', ',')
char_counts('review_title', ',')
char_counts('review_content', ',')
```

**We create splitting functions that attempts to split on commas, commas followed directly by a capital letter, and any commas not followed by spaces.**

### Create a User Table

## Exploratory Data Analysis


### Question: Are products with higher discounts more likely to have higher sales (approximated by rating_count)?


__A scatterplot is a good initial visualization here for seeing distribution and any correlations/patterns there may be.__

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(wdf['discount_percentage'], wdf['rating_count'], alpha=0.6, color='blue')

plt.xlabel('Discount Percentage (%)', fontsize=12)
plt.ylabel('Rating Count', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

#### Initial Observations
- Correlation seems low with this much noise
- We should apply additional constraints for testing


#### Break data into bins for discount percentage


```python
bins = [0, 20, 40, 60, 80, 100]
labels = ['0–20%', '21–40%', '41–60%', '61–80%', '81–100%']
wdf_obv_1 = wdf.copy()
# wdf_obv_1 = wdf_obv_1[wdf_obv_1['rating'] > 4.0]
wdf_obv_1['discount_range'] = pd.cut(wdf_obv_1['discount_percentage'], bins=bins, labels=labels)

summary = wdf_obv_1.groupby('discount_range', observed=True)['rating_count'].agg(['mean', 'count'])
summary['percent_increase'] = summary['mean'].pct_change() * 100
summary = summary.reset_index()

from IPython.display import display
print('\n')
display(summary)
```

**The percentage increase between discount brackets is unexpected (ignoring top and bottom brackets).**

#### Inspect actual price to be used as an additional constraint

```python
print(wdf['actual_price'].describe())
```

**Break data into (5) quantile price bins**

```python
# Create price bins based on quantiles
price_bins = [0] + list(wdf_obv_1['actual_price'].quantile([0.2, 0.4, 0.6, 0.8]).values) + [wdf_obv_1['actual_price'].max()]
price_labels = ['Quantile 1', 'Quantile 2', 'Quantile 3', 'Quantile 4', 'Quantile 5']

# Add a new column for price groups
wdf_obv_1['price_group'] = pd.cut(wdf_obv_1['actual_price'], bins=price_bins, labels=price_labels)

# Visualize data for all 5 groups
for group in price_labels:
    group_data = wdf_obv_1[wdf_obv_1['price_group'] == group]

    # Check if there is any data in the group
    if len(group_data) > 0:
        plt.figure(figsize=(8, 5))
        plt.scatter(group_data['discount_percentage'], group_data['rating_count'], alpha=0.6)
        plt.title(f'Actual Price {group}', fontsize=14)
        plt.xlabel('Discount Percentage (%)', fontsize=12)
        plt.ylabel('Rating Count', fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.show()
    else:
        print(f"No data available for {group}")
```
